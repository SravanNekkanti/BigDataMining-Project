```python
accidents_df1['Weather_Timestamp'] = pd.to_datetime(accidents_df1['Weather_Timestamp'])

accidents_df1['month'] = accidents_df1['Weather_Timestamp'].dt.month

accidents_df1['mean_temp_by_month'] = accidents_df1.groupby(['Zipcode', 'month'])['Temperature(F)'].transform('mean')
accidents_df1['mean_windchill_by_month'] = accidents_df1.groupby(['Zipcode', 'month'])['Wind_Chill(F)'].transform('mean')
accidents_df1['mean_humidity_by_month'] = accidents_df1.groupby(['Zipcode', 'month'])['Humidity(%)'].transform('mean')
accidents_df1['mean_pressure_by_month'] = accidents_df1.groupby(['Zipcode', 'month'])['Pressure(in)'].transform('mean')
accidents_df1['mean_visibility_by_month'] = accidents_df1.groupby(['Zipcode', 'month'])['Visibility(mi)'].transform('mean')
accidents_df1['mean_windspeed_by_month'] = accidents_df1.groupby(['Zipcode', 'month'])['Wind_Speed(mph)'].transform('mean')
accidents_df1['mean_precipitation_by_month'] = accidents_df1.groupby(['Zipcode', 'month'])['Precipitation(in)'].transform('mean')
```

```python
accidents_df1.columns
```

```python
accidents_df1['Temperature(F)'].fillna(accidents_df1['mean_temp_by_month'], inplace=True)
accidents_df1['Wind_Chill(F)'].fillna(accidents_df1['mean_windchill_by_month'], inplace=True)
accidents_df1['Humidity(%)'].fillna(accidents_df1['mean_humidity_by_month'], inplace=True)
accidents_df1['Pressure(in)'].fillna(accidents_df1['mean_pressure_by_month'], inplace=True)
accidents_df1['Visibility(mi)'].fillna(accidents_df1['mean_visibility_by_month'], inplace=True)
accidents_df1['Wind_Speed(mph)'].fillna(accidents_df1['mean_windspeed_by_month'], inplace=True)
accidents_df1['Precipitation(in)'].fillna(accidents_df1['mean_precipitation_by_month'], inplace=True)
```

```python
city_dataframes = {city: group for city, group in accidents_df1.groupby('City')}
for each_city in city_dataframes:
    print(each_city)
```

```python
city_record_counts = {}

for city, city_df in city_dataframes.items():
    record_count = len(city_df)
    city_record_counts[city] = record_count

city_record_counts_file = "city_record_counts.txt"

# You can print the record counts for each city:
with open(city_record_counts_file, 'w') as file:
    for city, count in city_record_counts.items():
        file.write(f"City: {city}, Record Count: {count}\n")
```

```python
major_cities_dataframes = {}
other_cities_dataframe = pd.DataFrame()

major_threshold = 10000

for city, city_df in city_dataframes.items():
    if len(city_df) > major_threshold:
        major_cities_dataframes[city] = city_df
    else:
        other_cities_dataframe = pd.concat([other_cities_dataframe, city_df])

major_cities_csv_filename = "major_cities.csv"
other_cities_csv_filename = "other_cities.csv"

for city, city_df in major_cities_dataframes.items():
    city_df.to_csv(f"{major_cities_csv_filename[:-4]}_{city}.csv", index=False)

other_cities_dataframe.to_csv(other_cities_csv_filename, index=False)

print(f"Major cities data saved to {major_cities_csv_filename}")
print(f"Other cities data saved to {other_cities_csv_filename}")
```

```python
charlotte_df = pd.read_csv("D:\\UIC\\Fall 2023 Project\\major_cities_Charlotte.csv")
charlotte_df.isnull().sum()
```

```python
charlotte_df.head()
```

```python
from haversine import haversine
import pandas as pd
import numpy as np

missing_temp_in_charlotte_df = charlotte_df[charlotte_df['Temperature(F)'].isna()]

for index, each_missing_entry in missing_temp_in_charlotte_df.iterrows():
    charlotte_df['Distance'] = charlotte_df.apply(lambda row: haversine((each_missing_entry['Start_Lat'], each_missing_entry['Start_Lng']), (row['Start_Lat'], row['Start_Lng'])), axis=1)
    
    nearby_entries = charlotte_df[charlotte_df['Distance'] <= 5]

    missing_date = pd.to_datetime(each_missing_entry['Start_Time'], format='mixed').date()

    three_day_entries = nearby_entries[pd.to_datetime(nearby_entries['Start_Time'], format='mixed').apply(lambda x: abs((x.date() - missing_date).days) <= 1)]

    if not three_day_entries.empty:
        average_temp = three_day_entries['Temperature(F)'].mean()
        charlotte_df.loc[index, 'Temperature(F)'] = average_temp
        
```

```python
charlotte_df.isnull().sum()
```

```python
from haversine import haversine
import pandas as pd
import numpy as np

missing_wind_chill_in_charlotte_df = charlotte_df[charlotte_df['Wind_Chill(F)'].isna()]

for index, each_missing_entry in missing_wind_chill_in_charlotte_df.iterrows():
    charlotte_df['Distance'] = charlotte_df.apply(lambda row: haversine((each_missing_entry['Start_Lat'], each_missing_entry['Start_Lng']), (row['Start_Lat'], row['Start_Lng'])), axis=1)
    
    nearby_entries = charlotte_df[charlotte_df['Distance'] <= 5]

    missing_date = pd.to_datetime(each_missing_entry['Start_Time'], format='mixed').date()

    three_day_entries = nearby_entries[pd.to_datetime(nearby_entries['Start_Time'], format='mixed').apply(lambda x: abs((x.date() - missing_date).days) <= 1)]

    if not three_day_entries.empty:
        average_wind_chill = three_day_entries['Wind_Chill(F)'].mean()
        charlotte_df.loc[index, 'Wind_Chill(F)'] = average_wind_chill
```
